```python
def is_palindrome(number):
    # Convert the number to a string
    num_str = str(number)
    # Reverse the string
    reversed_str = num_str[::-1]
    # Compare the original and reversed strings
    if num_str == reversed_str:
        return True
    else:
        return False

# Example usage:
num1 = 121
num2 = 12345
num3 = 9009

print(f"Is {num1} a palindrome? {is_palindrome(num1)}")
print(f"Is {num2} a palindrome? {is_palindrome(num2)}")
print(f"Is {num3} a palindrome? {is_palindrome(num3)}")

# You can also take input from the user:
# try:
#     user_input = int(input("Enter a number: "))
#     print(f"Is {user_input} a palindrome? {is_palindrome(user_input)}")
# except ValueError:
#     print("Invalid input. Please enter an integer.")
```
