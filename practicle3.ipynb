```python
import numpy as np

# Create a random 3x3 square matrix
matrix = np.random.randint(0, 10, size=(3, 3))

print("The 3x3 matrix is:")
print(matrix)

# Compute the determinant of the matrix
determinant = np.linalg.det(matrix)

print("\nThe determinant of the matrix is:")
print(determinant)
```

```python
import numpy as np

# Compute the transpose of the matrix
transposed_matrix = matrix.T

print("\nThe transposed matrix is:")
print(transposed_matrix)
```

```python
import numpy as np

# Define the coefficient matrix A
A = np.array([[2, 3], [4, 5]])

# Define the constant vector B
B = np.array([10, 20])

# Solve for x (where x = [x, y])
solution = np.linalg.solve(A, B)

print("The solution to the system of equations is:")
print(f"x = {solution[0]:.2f}")
print(f"y = {solution[1]:.2f}")
```
