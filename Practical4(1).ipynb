```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data from a normal distribution
mu, sigma = 0, 0.1 # mean and standard deviation
s = np.random.normal(mu, sigma, 1000)

# Plot the histogram
fig = plt.figure(figsize=(10, 6))
plt.hist(s, bins=50, density=True, alpha=0.6, color='g', label='Histogram')

# The following lines for 'best fit' and statistical measures have been removed
# to plot only the histogram as requested.

title = f"Histogram of Normally Distributed Data (mu = {mu:.2f}, sigma = {sigma:.2f})"
plt.title(title)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.legend()
plt.show()
```
