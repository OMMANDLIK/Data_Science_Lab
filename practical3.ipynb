```python
import numpy as np

# Create a random 3x3 matrix
matrix = np.random.randint(0, 10, size=(3, 3))

print("Created 3x3 Matrix:")
print(matrix)
```

```python
# Compute the determinant of the matrix
determinant = np.linalg.det(matrix)

print(f"\nDeterminant of the matrix: {determinant}")
```

```python
import numpy as np

# Define the matrix A (using the 3x3 matrix from Task 1)
A = np.array([[4, 1, 7],
              [6, 2, 5],
              [6, 8, 0]])

print("Original Matrix A:")
print(A)

# Compute the determinant to check for invertibility
determinant_A = np.linalg.det(A)

if abs(determinant_A) < 1e-9: # Check if determinant is close to zero
    print("\nThe matrix is singular, so its inverse cannot be computed.")
else:
    inverse_A = np.linalg.inv(A)
    print(f"\nDeterminant of A: {determinant_A}")
    print("Inverse of Matrix A:")
    print(inverse_A)
```
