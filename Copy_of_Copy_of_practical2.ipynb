First, let's define a graph using an adjacency list representation. We'll use a dictionary where keys are nodes and values are lists of their neighbors.

```python
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}

print("Defined Graph:")
for node, neighbors in graph.items():
    print(f"{node}: {neighbors}")
```

Next, here's the implementation for Breadth-First Search (BFS). BFS explores all of the neighbor nodes at the present depth prior to moving on to the nodes at the next depth level.

```python
from collections import deque

def bfs(graph, start_node):
    visited = set()
    queue = deque([start_node])
    bfs_traversal = []

    visited.add(start_node)

    while queue:
        current_node = queue.popleft()
        bfs_traversal.append(current_node)

        for neighbor in graph[current_node]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)

    return bfs_traversal

print(f"BFS Traversal starting from 'A': {bfs(graph, 'A')}")
```

Finally, here's the implementation for Depth-First Search (DFS). DFS explores as far as possible along each branch before backtracking.

```python
def dfs(graph, start_node):
    visited = set()
    stack = [start_node] # Using a list as a stack
    dfs_traversal = []

    while stack:
        current_node = stack.pop()
        if current_node not in visited:
            visited.add(current_node)
            dfs_traversal.append(current_node)

            # Add unvisited neighbors to the stack
            # We iterate in reverse to maintain typical DFS order for adjacent nodes
            for neighbor in reversed(graph[current_node]):
                if neighbor not in visited:
                    stack.append(neighbor)
    return dfs_traversal

print(f"DFS Traversal starting from 'A': {dfs(graph, 'A')}")
```
